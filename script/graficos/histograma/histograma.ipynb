# Adicionando anotações e linhas horizontais

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.set_title("Comprimento da Pétala", fontsize=20, pad=10)
eixo.grid(True)

eixo.hist(df['comprimento_pétala'], bins=20)

eixo.annotate(
    "$mu = {0:.2f}$\n$sigma = {1:.2f}$".format(mu,sigma),
    xy=(4.5,20),
    fontsize=20
)

eixo.axvline(mu, color='k', ls='--')
eixo.annotate('Média', xy=(mu-1.3,28), fontsize=20)

eixo.tick_params(labelsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma, mediana = [
                df['comprimento_pétala'].mean(),
                df['comprimento_pétala'].std(),
                df['comprimento_pétala'].median()
]

eixo.set_title("Comprimento da Pétala", fontsize=20, pad=10)
eixo.grid(True)

eixo.hist(df['comprimento_pétala'], bins=20)

eixo.annotate(
    "$mu = {0:.2f}$\n$sigma = {1:.2f}$".format(mu,sigma),
    xy=(4.5,20),
    fontsize=20
)

eixo.axvline(mu, color='k', ls='--')
eixo.annotate('Média', xy=(mu-1.3,28), fontsize=20)

eixo.axvline(mediana, color='g', ls='--')
eixo.annotate('Mediana', xy=(mediana,31), fontsize=20)

eixo.tick_params(labelsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

mu, sigma, mediana = [
                df['comprimento_pétala'].mean(),
                df['comprimento_pétala'].std(),
                df['comprimento_pétala'].median()
]

eixo.set_title("Comprimento da Pétala", fontsize=20, pad=10)
eixo.grid(True)

eixo.hist(df['comprimento_pétala'], bins=20)

eixo.annotate(
    "$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu,sigma),
    xy=(5,20),
    fontsize=20
)

eixo.axvline(mu, color='k', ls='--')
eixo.annotate('Média', xy=(mu-1.3,28), fontsize=20)

eixo.axvline(mediana, color='g', ls='--')
eixo.annotate('Mediana', xy=(mediana,31), fontsize=20, color='g')

eixo.tick_params(labelsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)

plt.show()
```

# Combinando e salvando figuras

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == "Iris-versicolor"]

mu, sigma, mediana = [
                df_iv['comprimento_pétala'].mean(),
                df_iv['comprimento_pétala'].std(),
                df_iv['comprimento_pétala'].median()
]

eixo.set_title("Iris Versicolor", fontsize=20, pad=10)
eixo.grid(True)

eixo.hist(df_iv['comprimento_pétala'], bins=20)

eixo.annotate(
    "$\mu = {0:.2f}$\n$\sigma = {1:.2f}$".format(mu,sigma),
    xy=(4.6,6),
    fontsize=20
)

eixo.axvline(mu, color='k', ls='--')
eixo.annotate(
    'Média',
    xytext=(3.5,5.5),
    fontsize=20,
    arrowprops=dict(facecolor='r', arrowstyle='->'),
    xy=(mu,5.5),
)

eixo.axvline(mediana, color='g', ls='--')
eixo.annotate(
    'Mediana',
    xytext=(3.5,6.5),
    fontsize=20,
    color='g',
    arrowprops=dict(facecolor='r', arrowstyle='->'),
    xy=(mediana,6.5),
)

eixo.tick_params(labelsize=15)
eixo.set_xlabel('Comprimento da Pétala', fontsize=15)
eixo.set_ylabel('N° observações', fontsize=15)

plt.show()
resultados = "./results/"
if not os.path.exists(resultados):
    os.makedirs(resultados)

fig.savefig(resultados+'histograma_iv.png', bbox_inches='tight', dpi=300)
```
