# Fazendo as primeiras visualizações

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

from util import *

# sns.set(rc={"figure.figsize": (20,8)}, font_scale=2)
```

```python
local_dados = "./../data/"
dic_list = os.listdir(local_dados)
dic_list = [local_dados+i for i in dic_list]
printLis(dic_list)
```

```python
df = pd.read_csv(dic_list[0])
df.head()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Gráfico de Dispersão", fontsize=20, pad=10)
eixo.scatter(x=df['comprimento_sépala'], y=df['largura_sépala'])

eixo.set_xlabel("Comprimento da Sépala", fontsize=15)
eixo.set_ylabel("Largura da Sépala", fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Gráfico de Dispersão", fontsize=20, pad=10)
eixo.scatter(x=df['comprimento_sépala'], y=df['largura_sépala'])

eixo.tick_params(labelsize=15)
eixo.set_xlabel("Comprimento da Sépala", fontsize=15)
eixo.set_ylabel("Largura da Sépala", fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {
    'Iris-setosa': 'r',
    'Iris-versicolor': 'b',
    'Iris-virginica': 'g',
}

for especie in df['espécie'].unique():
  tmp = df[df['espécie'] == especie]
  eixo.scatter(
      x=tmp['comprimento_sépala'],
      y=tmp['largura_sépala'],
      color=cores[especie]
  )

eixo.set_title("Gráfico de Dispersão", fontsize=20, pad=10)

eixo.tick_params(labelsize=15)
eixo.set_xlabel("Comprimento da Sépala", fontsize=15)
eixo.set_ylabel("Largura da Sépala", fontsize=15)

eixo.legend(cores, fontsize=20, loc='best')

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {
    'Iris-setosa': 'r',
    'Iris-versicolor': 'b',
    'Iris-virginica': 'g',
}

marcadores = {
    'Iris-setosa': 'x',
    'Iris-versicolor': 'o',
    'Iris-virginica': 'v',
}

for especie in df['espécie'].unique():
  tmp = df[df['espécie'] == especie]
  eixo.scatter(
      x=tmp['comprimento_sépala'],
      y=tmp['largura_sépala'],
      color=cores[especie],
      marker=marcadores[especie]
  )

eixo.set_title("Gráfico de Dispersão", fontsize=20, pad=10)

eixo.tick_params(labelsize=15)
eixo.set_xlabel("Comprimento da Sépala", fontsize=15)
eixo.set_ylabel("Largura da Sépala", fontsize=15)

eixo.legend(cores, fontsize=20, loc='best')

plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])

cores = {
    'Iris-setosa': 'r',
    'Iris-versicolor': 'b',
    'Iris-virginica': 'g',
}

marcadores = {
    'Iris-setosa': 'x',
    'Iris-versicolor': 'o',
    'Iris-virginica': 'v',
}

for especie in df['espécie'].unique():
  tmp = df[df['espécie'] == especie]
  eixo.scatter(
      x=tmp['comprimento_sépala'],
      y=tmp['largura_sépala'],
      color=cores[especie],
      marker=marcadores[especie],
      s=100
  )

eixo.set_title("Gráfico de Dispersão", fontsize=20, pad=10)

eixo.tick_params(labelsize=15)
eixo.set_xlabel("Comprimento da Sépala", fontsize=15)
eixo.set_ylabel("Largura da Sépala", fontsize=15)

eixo.legend(cores, fontsize=20, loc='best')

plt.show()
resultados = "./results/"
if not os.path.exists(resultados):
    os.makedirs(resultados)

fig.savefig(resultados+'scatter.png', bbox_inches='tight', dpi=300)
```
