# Fazendo as primeiras visualizações

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

from util import *

# sns.set(rc={"figure.figsize": (20,8)}, font_scale=2)
```

```python
local_dados = "./../data/"
dic_list = os.listdir(local_dados)
dic_list = [local_dados+i for i in dic_list]
printLis(dic_list)
```

```python
# monitoramento_tempo.csv
df = pd.read_csv(dic_list[1])
df.head()
```

```python
df.info()
```

```python
df['data'] = pd.to_datetime(df['data'])
df.info()
```

```python
temperatura_por_dia_da_semana = df.groupby("dia_da_semana")['temperatura'].mean()
nome_dias = ["Domingo","Segunda","Terça","Quarta","Quinta","Sexta","Sábado"]
temperatura_por_dia_da_semana = temperatura_por_dia_da_semana.loc[nome_dias]
temperatura_por_dia_da_semana
```

```python
fig = plt.figure(figsize=(5,4))

eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Temperatura por dia da Semana", fontsize=20, pad=10)
eixo.pie(temperatura_por_dia_da_semana, labels=temperatura_por_dia_da_semana.index)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))

eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Temperatura por dia da Semana", fontsize=20, pad=10)
eixo.pie(
    temperatura_por_dia_da_semana,
    labels=temperatura_por_dia_da_semana.index,
    autopct="%.1f%%"
)

plt.show()
```

```python
fig = plt.figure(figsize=(5,4))

eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

eixo.set_title("Temperatura por dia da Semana", fontsize=20, pad=10)
eixo.pie(
    temperatura_por_dia_da_semana,
    labels=temperatura_por_dia_da_semana.index,
    autopct="%.1f%%",
    explode=explodir
)

plt.show()
```

```python
fig = plt.figure(figsize=(7,6))

eixo = fig.add_axes([0,0,1,1])

explodir = [0.1,0,0,0,0,0.1,0.1]

eixo.set_title("Temperatura por dia da Semana", fontsize=20, pad=10)
eixo.pie(
    temperatura_por_dia_da_semana,
    labels=temperatura_por_dia_da_semana.index,
    autopct="%.1f%%",
    explode=explodir,
    shadow=True,
    textprops={'fontsize': 15}
)

plt.show()
resultados = "./results/"
if not os.path.exists(resultados):
    os.makedirs(resultados)

fig.savefig(resultados+'pizza.png', bbox_inches='tight', dpi=300)
```
