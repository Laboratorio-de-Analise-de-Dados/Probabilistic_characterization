```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# pd.options.plotting.backend = "plotly"

from util import *

# sns.set(rc={"figure.figsize": (20,8)}, font_scale=2)
```

```python
local_dados = "./../data/"
dic_list = os.listdir(local_dados)
dic_list = [local_dados+i for i in dic_list]
printLis(dic_list)
```

```python
# monitoramento_tempo.csv
df = pd.read_csv(dic_list[0])
df.head()
```

```python
fig = plt.figure(figsize=(5,4))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Largura da Pétala", fontsize=20, pad=10)

eixo.boxplot(df['largura_pétala'])

eixo.tick_params(labelsize=15)
eixo.set_xticklabels(['Largura Pétala'])

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Largura da Pétala", fontsize=20, pad=10)

eixo.boxplot(df.drop('espécie', axis=1).values)

eixo.tick_params(labelsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])

eixo.set_title("Largura da Pétala", fontsize=20, pad=10)

eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)

eixo.tick_params(labelsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])


eixo.set_title("Largura da Pétala", fontsize=20, pad=10)

cores = ['red','blue','orange','green']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)

for caixa, cor in zip(caixas['boxes'], cores):
  caixa.set(color=cor)


eixo.tick_params(labelsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
eixo = fig.add_axes([0,0,1,1])


eixo.set_title("Largura da Pétala", fontsize=20, pad=10)

cores = ['red','blue','orange','green']

caixas = eixo.boxplot(df.drop('espécie', axis=1).values, patch_artist=True)

for caixa, cor in zip(caixas['boxes'], cores):
  caixa.set(color=cor)

for outlier in caixas['fliers']:
  outlier.set(marker='x', markersize=8)
  
for median in caixas['medians']:
  median.set(color='k', lw=2)

eixo.tick_params(labelsize=15)
eixo.set_xticklabels(df.drop('espécie', axis=1).columns)

plt.show()
resultados = "./results/"
if not os.path.exists(resultados):
    os.makedirs(resultados)

fig.savefig(resultados+'boxplot.png', bbox_inches='tight', dpi=300)
```
